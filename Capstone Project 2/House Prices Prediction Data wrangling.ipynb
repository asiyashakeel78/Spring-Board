### 1.7  Check for missing Data<a id='1.7_Check_for_missing_Data'></a>

[<a href="#1.Data-wrangling">Back to top</a>]

```python
missing = pd.concat([df.isnull().sum(), 100 * df.isnull().mean()], axis=1)
missing.columns=['count', '%']
missing.sort_values(by='count' , ascending=False)
```

```python
missing = df.isnull().sum()
missing = missing[missing > 0]
missing.sort_values(inplace=True)
ax=missing.plot.bar()
ax.bar_label(ax.containers[0])
ax.set_title('Distribution of missing values', color='blue')
ax.set_xlabel('Features with Missing Values', color='blue')
ax.set_ylabel('Frequency', color='blue')

plt.show()
```

19 attributes have missing values, 5 over 50% of all data. Most of times NA means lack of subject described by attribute, like missing pool, fence, no garage and basement.

```python
df.isnull().sum()
```

__<font color=blue> Separate the Numerical and Categorical features:__

```python
numeric_cols = df.select_dtypes(include=np.number).columns.to_list()
```

```python
categorical_cols = df.select_dtypes('object').columns.to_list()
```

```python
ax= sns.histplot(df["SalePrice"], kde=True, bins=50)
fig = plt.gcf()
fig.set_size_inches(9, 5)
ax.set_title("Distribution of Sale Prices")
ax.set_ylabel('No of Houses')
```

```python
counts = df['YrSold'].value_counts()
colors = plt.cm.Set3(np.linspace(0, 1, len(counts))) 

counts.plot(kind='bar', color=colors)

for i, count in enumerate(counts):
    plt.text(i, count+1, str(count), ha='center', va='bottom')
    
plt.xlabel('Year Sold')
plt.ylabel('No of Houses')
plt.xticks(rotation=45)

total_count = counts.sum()

Title= f'Over the Years {total_count} houses were sold'

plt.title(Title)

# display chart
plt.show()
```

__<font color=blue> Average Saleprice as per Neighborhood__

```python
Neighborhood_price_means = df.groupby('Neighborhood')[['SalePrice']].mean().sort_values(by='SalePrice', ascending=False)
Neighborhood_price_means
```

```python


Neighborhood_price_means.plot(kind='bar', color=colors)
    
plt.xlabel('Neighborhood')
plt.ylabel('Average Saleprice')

Title= 'Average Saleprice as per Neighborhood'

plt.title(Title)

# display chart
plt.show()
```
