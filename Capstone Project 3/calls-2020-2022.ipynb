```python
monthly_calls = df1.resample('M').sum()
```

```python
monthly_calls.head() 
```

```python

import matplotlib.dates as mdates
```

```python
plt.figure(figsize=(10, 6))
plt.plot(monthly_calls.index, monthly_calls["Call count"], marker='o', linestyle='-')
plt.xlabel("Date")
plt.ylabel("Monthly Calls")
plt.title("Monthly Calls over Time")
plt.grid(True)
plt.xticks(rotation=45)

# Show months on the x-axis explicitly
plt.gca().xaxis.set_major_locator(mdates.MonthLocator())
plt.gca().xaxis.set_major_formatter(mdates.DateFormatter('%b %Y'))

plt.tight_layout()
plt.show()
```
