### Missing values

```python
missing = pd.concat([df.isnull().sum(), 100 * df.isnull().mean()], axis=1)
missing.columns=['count', '%']
missing.sort_values(by='count' , ascending=False)
```

```python
threshold = 10
columns_to_plot = missing[missing['%'] >= threshold]

# Sort the columns_to_plot DataFrame in descending order based on '%'
columns_to_plot = columns_to_plot.sort_values(by='%', ascending=False)

# Plot the missing values greater than 50%
plt.figure(figsize=(12, 6))
bars = plt.bar(columns_to_plot.index, columns_to_plot['%'])
plt.xlabel('Columns')
plt.ylabel('Missing Percentage')
plt.title('Features with Missing Values ')
plt.xticks(rotation=90)  # To rotate the column names for better visibility

for bar in bars:
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width() / 2, height, f'{height:.2f}%', ha='center', va='bottom')

plt.tight_layout()
plt.show()
```

### Check yearly calls 

```python
yearly_calls= df1.groupby(by="Created Date")["Call count"].sum().reset_index()

yearly_calls["Created Date"] = pd.to_datetime(yearly_calls["Created Date"])

yearly_calls.set_index("Created Date", inplace=True)

yearly_calls = yearly_calls.resample('y').sum()

yearly_calls.head()
```

```python
yearly_calls.describe()
```

```python
plt.figure(figsize=(12, 6))
plt.plot(yearly_calls.index, yearly_calls['Call count'], marker='o', linestyle='-')
plt.xlabel("Year")
plt.ylabel("Yearly Call Count")
plt.title("Yearly Call Counts over Time")
plt.grid(True)
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
Agency_calls = df[['Created Date', 'Agency']]
```

```python
agency_counts = df['Agency'].value_counts()

# Filter the 'Agency' based on the condition where the total count is greater than 104603
selected_agencies = agency_counts[agency_counts > 104603].index

# Filter the DataFrame to keep only the selected agencies
df_selected = df[df['Agency'].isin(selected_agencies)]

# Convert 'Created Date' to datetime format
df_selected['Created Date'] = pd.to_datetime(df_selected['Created Date'])

# Extract the year from 'Created Date'
df_selected['Year'] = df_selected['Created Date'].dt.year

# Group the data by year and 'Agency', and unstack it for a grouped bar plot
grouped_data = df_selected.groupby(['Year', 'Agency']).size().unstack()

# Create a grouped bar plot
plt.figure(figsize=(12, 6))
grouped_data.plot(kind='bar', stacked=True)
plt.xlabel('Year')
plt.ylabel('Count')
plt.title('Agency Calls for Each Year')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.legend(title='Agency', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```

# Top 10 Agencies

```python
Selected_agencies = Agency_calls[Agency_calls['Agency'].isin(Agency_counts[Agency_counts > 100000].index)]
```

```python
Selected_agencies ['Call count'] = 1
```

```python
Selected_agencies.head()
```

```python
grouped_data = Selected_agencies.groupby(['Agency', 'Created Date'])['Call count'].sum().reset_index()

# Sort the data in descending order based on 'Call count' for each agency feature
sorted_data = grouped_data.sort_values(by='Call count', ascending=False)

# Iterate over each agency feature and create separate plots
for agency_feature in sorted_data['Agency'].unique():
    agency_data = sorted_data[sorted_data['Agency'] == agency_feature]
    agency_data.set_index('Created Date', inplace=True)

    # Create the plot
    ax = agency_data.plot(figsize=(15, 6))

    # Set the title for each plot based on the agency feature
    ax.set_title(f'Number of Calls per Day for {agency_feature} Agency')

    # Optionally, you can set the labels for the x and y axes
    ax.set_xlabel('Date')
    ax.set_ylabel('Call Count')

    # Show the plot for each agency feature
    plt.show()
```

```python

import matplotlib.cm as cm

df['Created Date'] = pd.to_datetime(df['Created Date'])

# Extract the year from the "Created Date" column
df['Year'] = df['Created Date'].dt.year

# Loop through each Borough and create separate plots
for borough in df['Borough'].unique():
    # Filter the DataFrame for the current Borough
    borough_df = df[df['Borough'] == borough]
    
    # Group by year and agency and count the occurrences
    grouped_df = borough_df.groupby(['Year', 'Agency']).size().unstack(fill_value=0)
    
    # Calculate the color map based on the counts of each Agency
    cmap = cm.get_cmap('Dark2', len(grouped_df.columns))
    
    # Create the stacked bar plot
    fig, ax = plt.subplots(figsize=(8, 5))
    grouped_df.plot(kind='bar', stacked=True, ax=ax, cmap=cmap)
    
    # Customize the plot
    ax.set_xlabel('Year')
    ax.set_ylabel('Count')
    ax.set_title(f'Number of calls by Agency for {borough}')
    ax.legend(title='Agency', loc='upper left', bbox_to_anchor=(1, 1))
    
    # Show the plot
    plt.tight_layout()
    plt.show()
```
